```python
from datasets import load_dataset
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from scipy import spatial
import nltk
import torch
from sentence_transformers import SentenceTransformer, util

# Download NLTK stopwords
nltk.download('stopwords')

# Load dataset
dataset = load_dataset('ms_marco', 'v1.1')

corpus = []
subset = dataset['test']

# Extract data from dataset
for sample in subset:
    query_type = sample.get('query_type', None)
    if query_type != 'entity':
        continue
    query_id = sample.get('query_id', None)
    query_str = sample.get('query', None)
    passages_dict = sample.get('passages', None)
    if passages_dict:
        passages_text_lst = [passage.get('passage_text', '') for passage in passages_dict if passage.get('is_selected', 0) == 1]
        corpus += passages_text_lst

# Ensure the corpus is not empty
if len(corpus) == 0:
    raise ValueError("The corpus is empty. Please check the dataset or filtering criteria.")

# Load SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode corpus to obtain embeddings
corpus_embeddings = model.encode(corpus, convert_to_tensor=True)

# Define cosine similarity function
def similarity(query_embeddings, corpus_embeddings):
    return util.cos_sim(query_embeddings, corpus_embeddings)

# Rank documents by similarity to query
def ranking(query, top_k=10):
    query_embeddings = model.encode(query, convert_to_tensor=True)
    cos_scores = similarity(query_embeddings, corpus_embeddings)
    top_results = torch.topk(cos_scores, k=min(top_k, len(cos_scores[0])))
    return top_results

# Query and ranking
query_list = ['what is the official language in Fiji']
top_k = 10
for query in query_list:
    top_results = ranking(query, top_k)
    print(f'Query: {query}')
    print('=== Relevant docs ===')
    for idx in range(len(top_results.values)):
        score = top_results.values[idx].item()
        doc_idx = top_results.indices[idx].item()

        print(f'Document rank {idx + 1}:')
        print(corpus[doc_idx], f'\n(Score: {score:.4f})', '\n')
```
